```python
def ct_func(ct):
    if ct =='ADLD':
        return 1
    else:
        return 2
```

```python
def cpcfunc(cat):
    if cat == 'A':
        return 1
    elif cat =='B':
        return 2
    elif cat == 'C':
        return 3
    elif cat == 'D':
        return 4
    else:
        return 0
```

```python
train_data['claim_type'] = train_data['claimtype'].apply(ct_func)
test_data['claim_type'] = test_data['claimtype'].apply(ct_func)
```

```python
train_data['pincode_category'] =train_data['custpincodecategory'].apply(cpcfunc)
test_data['pincode_category'] =test_data['custpincodecategory'].apply(cpcfunc)
```
